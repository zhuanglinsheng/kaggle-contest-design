```python
import numpy as np
from sklearn.neighbors import KernelDensity
import matplotlib.pyplot as plt

events = np.array([1.2, 3.5, 4.7, 6.1, 8.9]).reshape(-1, 1)
ticks = np.linspace(0, 10, 1000).reshape(-1, 1)

kde = KernelDensity(bandwidth=0.5, kernel='gaussian')
kde.fit(events)
kde_ticks_eval = kde.score_samples(ticks)
intensity = np.exp(kde_ticks_eval) * len(events)

plt.plot(ticks, intensity)
plt.xlabel('Time')
plt.ylabel('Intensity')
plt.title('Nonparametric NHPP Intensity via KDE')
plt.show()
```
